```python
import numpy as np

# We have 100 cities, 955 connections between them, and 73 euros for fuel

start_cities = []
end_cities = []
distance = [[0 for row in range(0,100)] for col in range(0,100)]
fuel_costs = [[0 for row in range(0,100)] for col in range(0,100)]

distance_flat = []
fuel_costs_flat = []

#Parse instance file to create location, fuel and distances arrays
f = open('instance day 2 modified.txt', 'r')

# iterate over the lines in the file
for line in f:
    # split the line into a list of column values
    columns = line.strip().split(' ')
    # clean any whitespace off the items
    columns = [col.strip() for col in columns]

    start_cities.append(int(columns[0]))
    end_cities.append(int(columns[1]))

    print(columns)
    distance[int(columns[0]) - 1][int(columns[1]) - 1] = int(columns[2])
    
    
    
    fuel_costs[int(columns[0])- 1 ][int(columns[1]) - 1] = int(columns[3])



for i in distance:
        for j in i:
            distance_flat.append(j)    
for i in fuel_costs:
        for j in i:
            fuel_costs_flat.append(j)    


distance_flat = distance_flat[0:len(distance_flat)]

fuel_costs_flat = fuel_costs_flat[0:len(fuel_costs_flat)]




f.close()

# Write minizinc variables to a new file
g = open("day 2 variables v4.txt", "w")

g.write(f"start_cities = {start_cities};\n\n")
g.write(f"end_cities = {end_cities};\n\n")
g.write(f"distance = {distance_flat};\n\n")
g.write(f"fuel_costs = {fuel_costs_flat};\n\n")

g.close()

print(start_cities)
print(end_cities)
print(distance)
print(fuel_costs)
```
